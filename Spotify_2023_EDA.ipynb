## Number of Tracks Released Per Year ##

```python
df_spotify['released_year'].value_counts().sort_index().plot()

plt.title('Tracks Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')

plt.show()
```

```python
monthly_releases = df_spotify['released_month'].value_counts().sort_index()

monthly_releases.plot(kind='bar', color='blue', figsize=(10, 5), title='Tracks Released Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Tracks')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

print(f"Month with the most releases: {monthly_releases.idxmax()} ({monthly_releases.max()} tracks)")
```

## Attributes that seem to Influence Streams the Most ##

```python
correlation = df_spotify[['streams', 'bpm', 'danceability_%', 'energy_%']].corr()

print(correlation['streams'])
```

## Correlation between danceability_% and energy_%, and between valence_% and acousticness_% ##

```python
dance_energy_corr = df_spotify[['danceability_%', 'energy_%']].corr().iloc[0, 1]

# Calculate correlation for valence_% and acousticness_%
valence_acoustic_corr = df_spotify[['valence_%', 'acousticness_%']].corr().iloc[0, 1]

# Print the results
print(f"Correlation between danceability_% and energy_%: {dance_energy_corr}")
print(f"Correlation between valence_% and acousticness_%: {valence_acoustic_corr}")
```

## Analyzing the mean and median of tracks across Spotify and Apple Music ##

```python
spotify_playlists_mean = df_spotify['in_spotify_playlists'].mean()
spotify_charts_mean = df_spotify['in_spotify_charts'].mean()
apple_playlists_mean = df_spotify['in_apple_playlists'].mean()

spotify_playlists_median = df_spotify['in_spotify_playlists'].median()
spotify_charts_median = df_spotify['in_spotify_charts'].median()
apple_playlists_median = df_spotify['in_apple_playlists'].median()

# Print the results
print(f"Spotify Playlists - Mean: {spotify_playlists_mean}, Median: {spotify_playlists_median}")
print(f"Spotify Charts - Mean: {spotify_charts_mean}, Median: {spotify_charts_median}")
print(f"Apple Playlists - Mean: {apple_playlists_mean}, Median: {apple_playlists_median}")
```
